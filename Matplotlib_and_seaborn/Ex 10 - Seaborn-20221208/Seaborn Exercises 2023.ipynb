**1. sns.jointplot** - pay attention to the ranges on the plot

```python
sns.jointplot(x='fare',y='age',data=titanic,kind='scatter', xlim = (-100,600))
```

**2. sns.displot**

```python
sns.displot(titanic['fare'], kde = False, bins = 30)
```

**3. sns.boxplot**

```python
sns.boxplot(x="class", y="age", data=titanic,palette='rainbow')
```

**4. sns.countplot**

```python
sns.countplot(x='sex',data=titanic)
```

**5. sns.swarmplot**

```python
#Ignore the warnings:) 
sns.swarmplot(x="class", y="age", data=titanic)

```

**6. sns.heatmap**

```python
titanic.corr()
sns.heatmap(titanic.corr(), cmap = 'coolwarm')
```

**7. sns.FacetGrid**

```python
g = sns.FacetGrid(titanic, col="sex")
g.map(plt.hist, 'age')
```

 **8. sns.countplot**

```python
count_grath= sns.countplot(x='pclass',hue='survived', data=titanic, palette = 'Set1')
count_grath
```
