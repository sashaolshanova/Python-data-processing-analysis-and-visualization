**11. We would now like to determine the percentage of every the total invoice For each CustomerID.  We will do this in two ways: using merging and using transform:**

**a.** First, create a new column with the total price for each item purchased (based on quantity and unit price)

```python
retail['Total Price'] = retail['Quantity']*retail['UnitPrice']
retail
```

**b.** Find the total price of each invoice using grouby.

```python
order_total = retail.groupby('InvoiceNo')['Total Price'].sum()
order_total
```

**c.** Change the name of the Series to "Order Total" and then reset the index of the new data (so you will get a DataFrame). 

```python
new_name = order_total.rename("Order Total")
#new_name
order_total_data = pd.DataFrame(new_name.reset_index())
order_total_data
```

**d.** Use Merge function to combine the new data with the original data to a new DataFrame called "new_retail". <br>Add a new column with the percentage of the order named Percentage of Order.

```python
new_retail = retail.merge(order_total_data)
new_retail['Perentage of Order'] = new_retail['Total Price'] / new_retail['Order Total'] 
new_retail.head()
```

**The second approach is to use groupby along with transform.  This avoids the issue of reordering the index to obtain the new dataframe.  We retain the same number of items as in the original data set.**<br>
**e.** Use groupby and transform function to make a new column - Order Total.

```python
retail["Order Total"] = retail.groupby('InvoiceNo')['Total Price'].transform('sum')
retail
```

**f.** Add a column to with the percentage of the order.

```python
retail['Perentage of Order'] = retail['Total Price'] / retail['Order Total'] 
retail
```
