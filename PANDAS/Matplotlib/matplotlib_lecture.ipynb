```python
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np
mpl.__version__
```

```python
# Functional interface
plt.figure()
plt.subplot(2, 1, 1) # (rows, columns, panel number) 
x = np.linspace(-2*np.pi, 2*np.pi, 100)
plt.plot(x, np.sin(x)) 
# create the second panel and set current axis 
plt.subplot(2, 1, 2)
plt.plot(x, np.cos(x))
```

```python
# Object oriented interface
fig, ax = plt.subplots(2)
x = np.linspace(-2*np.pi, 2*np.pi, 100)
ax[0].plot(x, np.sin(x)) 
ax[1].plot(x, np.cos(x))
```

```python
# Line plots
fig = plt.figure()
ax = plt.axes()
print(type(fig), type(ax))
x = np.linspace(0, 10, 1000)
ax.plot(x, np.sin(x));
# Alternatively
plt.plot(x, np.sin(x))
```

```python
# Line color
x = np.linspace(-2*np.pi, 2*np.pi, 100)

plt.plot(x, np.sin(x - 0), color='blue') # specify color by name
plt.plot(x, np.sin(x - 1), color='g') # short color code (rgbcmyk)
plt.plot(x, np.sin(x - 2), color='0.75') # Grayscale between 0 and 1 
plt.plot(x, np.sin(x - 3), color='#FFDD44') # Hex code (RRGGBB from 00 to FF) 
plt.plot(x, np.sin(x - 4), color=(1.0,0.2,0.3)) # RGB tuple, values 0 to 1 
plt.plot(x, np.sin(x - 5), color='chartreuse'); # HTML color names
```

```python
# Line style
plt.plot(x, x + 0, linestyle='solid') 
plt.plot(x, x + 1, linestyle='dashed')
plt.plot(x, x + 2, linestyle='dashdot') 
plt.plot(x, x + 3, linestyle='dotted') 
# For short, you can use the following codes: 
plt.plot(x, x + 4, linestyle='-') # solid 
plt.plot(x, x + 5, linestyle='--') # dashed 
plt.plot(x, x + 6, linestyle='-.') # dashdot 
plt.plot(x, x + 7, linestyle=':') # dotted
```

```python
# Axes limits
plt.plot(x, np.sin(x)) 
plt.xlim(-2*np.pi - 0.2, 2*np.pi + 0.2) 
plt.ylim(-1.5, 1.5)
# Alternatively
#plt.axis([-2*np.pi - 0.2, 2*np.pi + 0.2, -1.5, 1.5])
```

```python
# Labeling plots
plt.plot(x, np.sin(x)) 
plt.xlim(-2*np.pi - 0.2, 2*np.pi + 0.2) 
plt.ylim(-1.5, 1.5)
plt.title("A Sine Curve") 
plt.xlabel("x") 
plt.ylabel("sin(x)")
```

```python
# Legends
plt.xlim(-2*np.pi - 0.2, 2*np.pi + 0.2) 
plt.ylim(-1.5, 1.5)
# Add titles & labels
plt.title("A Sine Curve") 
plt.xlabel("x") 
plt.ylabel("sin(x)")
# Add legends
plt.plot(x, np.sin(x), '-g', label='sin(x)')
plt.plot(x, np.cos(x), ':b', label='cos(x)') 
# one unit in x is equal to one unit in y 
plt.legend()
```

```python
# Set all properties at once
ax = plt.axes()
ax.plot(x, np.sin(x))
ax.set(xlim=(-2*np.pi - 0.2, 2*np.pi + 0.2), ylim=(-2, 2), xlabel='x', ylabel='sin(x)', title='A Simple Plot')
```

```python
# Contour plots
def f(x, y):
    return np.sin(x) ** 10 + np.cos(10 + y * x) * np.cos(x)

x = np.linspace(0, 5, 50)
y = np.linspace(0, 5, 40)

X, Y = np.meshgrid(x, y) # Create 2D grid
Z = f(X, Y)
plt.contour(X, Y, Z, colors='black')
plt.show()
plt.contour(X, Y, Z, 20, cmap='coolwarm')
plt.show()
plt.contourf(X, Y, Z, 20, cmap='coolwarm')
plt.colorbar()
plt.show()
```

```python
contours = plt.contour(X, Y, Z, 3, colors='black')
plt.clabel(contours, inline=True, fontsize=8)
plt.imshow(Z, extent=[0, 5, 0, 5], origin='lower', cmap='coolwarm', alpha=0.5)
plt.colorbar()
```

```python
# Histograms
import matplotlib.style  as st 

mu, sigma = 0, 0.1 # mean and st. dev.
s = np.random.normal(mu, sigma, 1000)
nbins = 30
st.use('ggplot') # return grid
plt.figure(figsize = (16,12)) 
count, bins, ignored = plt.hist(s, nbins, density=True, color='olive')
plt.plot(bins, 1/(sigma * np.sqrt(2 * np.pi)) *
         np.exp( - (bins - mu)**2 / (2 * sigma**2) ),
         linewidth=2, color='darkgreen')
```

```python
import pandas as pd

df = pd.read_csv('/Users/majort/Documents/Research/BIU/Courses/Python-data processing, analysis and visualization/Data Sets/glass.csv', encoding='cp1252')
elements = ['Na', 'Mg', 'Al', 'Si', 'Ca']
df[elements[:]] = df[elements[:]].replace(0.0, np.nan) 
print(df[elements[:]])

fig, axes = plt.subplots(1, 5, figsize=(17.5,2.5), dpi=600, sharex=False, sharey=True)
colors = ['tab:red', 'tab:blue', 'tab:green', 'tab:pink', 'tab:olive']

for i in range(len(elements)):
    na_data = df[elements[i]].values
    axes[i].hist(na_data, alpha=0.5, bins=100, density=True, stacked=True, label=elements[i], color=colors[i])
    axes[i].set_title(elements[i])
    axes[i].xaxis.set_major_formatter('{x:.1f}')
    axes[i].yaxis.set_major_formatter('{x:.2f}')
axes[0].set_ylabel("Frequency")
axes[2].set_xlabel("Percentage (%)")

plt.suptitle('Probability Histogram of Glass Elements', y=1.05, size=16)
fig.savefig('glass_elements.png', dpi=600, bbox_inches='tight')
```
